```python
import pandas as pd

neural_df = pd.read_csv("rgb-v5_eval_results.csv")
scobot_df = pd.read_csv("scobots_eval_results.csv")
iscobot_df = pd.read_csv("iscobots_eval_results.csv")
skiing_df = pd.read_csv("skiscobots_eval_results.csv")
abl_norel_df = pd.read_csv("abl_norel_eval_results.csv")
abl_onlyrel_df = pd.read_csv("abl_onlyrel_eval_results.csv")

iscobot_df = iscobot_df.drop(iscobot_df[(iscobot_df.env == "Skiing" )].index)
iscobot_df[iscobot_df["env"] == "Skiing"].sort_values(by=["reward_mean"])
all_df = pd.concat([neural_df, scobot_df, iscobot_df, skiing_df, abl_norel_df, abl_onlyrel_df])
abl_df = pd.concat([scobot_df,abl_norel_df, abl_onlyrel_df])
abl_df
```

```python
grouped = all_df.groupby(["env", "variant"])

agg_df = grouped["reward_mean"].agg(["mean", "std"])
agg_df
```

```python
agg_df.index = agg_df.index.set_levels(agg_df.index.levels[1].str.replace('rgbv5', 'neural'), level=1)
agg_df.index = agg_df.index.set_levels(agg_df.index.levels[1].str.replace('abl_norel', 'scobots_properties'), level=1)
agg_df.index = agg_df.index.set_levels(agg_df.index.levels[1].str.replace('abl_onlyrel', 'scobots_relations'), level=1)
agg_df.to_csv("abl_agg_results.csv")
```
