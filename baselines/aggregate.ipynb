```python
import pandas as pd

df = pd.read_csv("rgb-v5_eval_results.csv")
#df[df["env"] == "Skiing"].sort_values(by=["reward_mean"])
df
#df = df.drop(df[(df.score < 50) & (df.score > 20)].index)
#df[df["env"] == "Pong"]
```

```python
grouped = df.groupby(["env", "variant"])

agg_df = grouped["reward_mean"].agg(["mean", "std"])
agg_df
```

```python
agg_df.index = agg_df.index.set_levels(agg_df.index.levels[1].str.replace('rgbv4', 'original'), level=1)
agg_df.to_csv("agg_results.csv")
```
